## 4. Add a new column (15pts)

* Add a new column **Customer_Days** to store the total number of days each customer has been with the company. You may follow the steps below:
  * use `pd.to_datetime(df["Dt_Customer"])` to convert the string variable to a *date-time* variable

  * use `today = datetime.today()` to find the date of today

  * use `(today - df["Dt_Customer"]).dt.days` to find the difference between today and the onbard date.

```python
from datetime import datetime

# Converting the existing column to Datetime
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format = 'mixed', dayfirst = True)
#Saving the current date and time
today = datetime.today()
#Adding the new column with the difference (in days)
df["Customer_Days"] = (today - df["Dt_Customer"]).dt.days
display(df.head())
```

## 6. Numeric Variables (15pts)

* For `Income`, `Kidhome`, and `Teenhome`,
  * use boxplots to visualize their distributions.

* Based on these distributions, what can you say about the typical household profile of Nata's customers?

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
df['Income'].plot(kind='box', title='Income distribution')
plt.figure(figsize=(10,6))
df['Kidhome'].plot(kind='box', title='#Kids in household distribution')
plt.figure(figsize=(10,6))
df['Teenhome'].plot(kind='box', title='#Teens in household distribution')
```

<font color='red'>
The first thing we can notice is that the boxplots for Teenhome and Kidhome are the same, indicating that the distribution for the amount of kids and teens in the household are the same.
The mean and lower quartile are 0, with a maximum of 2. This means the majority of the households have 0 or at most 1 kid/teen, with a few having 2 children/teens at home. More precisely, 50% do not have kids/teens, 75% have at most one kid/teen and 100% have at most two. We can safely infer 50% don't have kids, 25% have one and 25% have two, seeing fractional values are not possible.

<font color='red'>
<br> Income-wise, the majority (50%) of the customers have an income between 25k and 75k, with a median around 50k.
